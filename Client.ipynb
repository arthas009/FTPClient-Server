```python
from ftplib import FTP
from tkinter import *
from tkinter import filedialog
import os
import glob
import io
from io import StringIO
import shutil

#We'll handle local operations with 'os' library
os.chdir("C:/localdizin")

print("done")
```

```python
#Create panel with 'tk' library
root = Tk()
root.geometry("1350x700+300+150")
root.resizable(width=True, height=True)

#Connect localhost:1656
ftp = FTP('')
ftp.connect('localhost',1656)
ftp.login(user='user', passwd='12345')
ftp.cwd('.') #replace with your directory


print("done")
```

```python
#RETR command retrieves file to local Computer. 
def downloadFile():
    targerfilename = 'C:/localdizin/'+str(download_filename_entry.get()) 
    localfile = open(targerfilename, 'wb')
    ftp.retrbinary('RETR '+str(download_filename_entry.get()), localfile.write, 1656)
    localfile.close()
    
    T2.delete('1.0', END)
    for file in os.listdir(os.getcwd()):
        T2.insert(END,str(file)+"\n")
      
    #ftp.quit() 

# Change working directory on remote server
def change_working_directory():
    ftp.cwd('./'+e2.get())
    T.delete('1.0', END)
    for i in ftp.nlst():
        T.insert(END,str(i)+"\n")

# Go back directory on remote server.
def go_back():
    ftp.cwd('../')
    T.delete('1.0', END)
    for i in ftp.nlst():
        T.insert(END,str(i)+"\n")

# Create file on remote server. STOR command stores a file on remote server
def createFileRemote():
    to_bytes = bytes(create_file_remote_input_text.get(), "UTF-8")
    textToWrite = io.BytesIO(to_bytes)
    ftp.storbinary('STOR '+create_file_remote.get(),textToWrite)
    T.delete('1.0', END)
    for i in ftp.nlst():
        T.insert(END,str(i)+"\n")

# Create Directory on Remote Server
def createDirectoryOnRemoteServer():
    ftp.mkd(create_directory_on_remote_server_directory_name.get())
    T.delete('1.0', END)
    for i in ftp.nlst():
        T.insert(END,str(i)+"\n")
        
# Delete file on remote server        
def deleteFileRemote():
    ftp.delete(delete_filename_entry.get())
    T.delete('1.0', END)
    for i in ftp.nlst():
        T.insert(END,str(i)+"\n")
# Delete Directory on Remote Server
def deleteDirectoryRemote():
    ftp.rmd(delete_directory_entry.get())
    T.delete('1.0', END)
    for i in ftp.nlst():
        T.insert(END,str(i)+"\n")
# Rename file on remote server        
def renameFile():
    ftp.rename(rename_file_remote_first.get(), rename_file_remote_last.get())
    T.delete('1.0', END)
    for i in ftp.nlst():
        T.insert(END,str(i)+"\n")
        
###########################################################################################################    
# Change working directory on local side
def change_working_directory_local():
    os.chdir('./'+e4.get())
    T2.delete('1.0', END)
    for file in os.listdir(os.getcwd()):
        T2.insert(END,str(file)+"\n")
# Go back directory on local side
def go_back_local():
    if os.getcwd() != ("C:\localdizin"):
        os.chdir('../')
    T2.delete('1.0', END)
    for file in os.listdir(os.getcwd()):
        T2.insert(END,str(file)+"\n")

# Upload a file from local to remote server using STOR command
def uploadFile():
    filename = download_filename_entry2.get() #replace with your file in your home folder
    ftp.storbinary('STOR '+filename, open(os.getcwd()+"/"+filename, 'rb'))
    T.delete('1.0', END)
    for i in ftp.nlst():
        T.insert(END,str(i)+"\n")

# Create file on local side
def createFileLocal():
    file = open(create_file_local.get(), "w+") 
    file.write(create_file_local_input_text.get()) 
    file.close()
    T2.delete('1.0', END)
    for file in os.listdir(os.getcwd()):
        T2.insert(END,str(file)+"\n")

# delete file on local side
def deleteFileLocal():
    os.remove(delete_filename_entry_local.get())
    T2.delete('1.0', END)
    for file in os.listdir(os.getcwd()):
        T2.insert(END,str(file)+"\n")

# delete directory on local side
def deleteDirectoryLocal():
    dirname = os.getcwd()+"\\"+delete_directory_entry_local.get()
    dirname=dirname.replace('\\','/')
    shutil.rmtree(dirname)
    T2.delete('1.0', END)
    for file in os.listdir(os.getcwd()):
        T2.insert(END,str(file)+"\n")

# rename File on local side
def renameFileLocal():
    os.rename(rename_file_local_first.get(),rename_file_local_last.get())
    T2.delete('1.0', END)
    for file in os.listdir(os.getcwd()):
        T2.insert(END,str(file)+"\n")    
        
# create Directory on local side
def createDirectoryOnLocal():
    os.mkdir(create_directory_on_local_directory_name.get())
    T2.delete('1.0', END)
    for file in os.listdir(os.getcwd()):
        T2.insert(END,str(file)+"\n")   

    
```
